```python
df.loc[df['Customer Zipcode'].isna(),['Customer Zipcode','Customer City']]
df.loc[df['Customer City']=="CA"][['Customer Zipcode']]
df['Customer Zipcode']=df['Customer Zipcode'].fillna(df['Customer Zipcode'].mode()[0])
```

```python
df.isna().sum()
df['Customer Lname'] = df['Customer Lname'].fillna(df['Customer Lname'].mode()[0])
df.isna().sum()
```

```python
df.dtypes
```

```python
str_columns = df.select_dtypes(include=['object']).columns
df[str_columns] = df[str_columns].apply(lambda x: x.str.strip())
```

```python
for col in str_columns:
    df[col] = df[col].str.encode('latin1',errors='ignore').str.decode('utf-8',errors='ignore')

df.dtypes
```

```python
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'], errors='coerce')
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'], errors='coerce')

df['Order State'] = df['Order State'].replace('Rajastn', 'Rajasthan')
df['Order State'] = df['Order State'].replace('Tokio', 'Tokyo')

df.groupby('Order State')['Order State'].count()
df['Order City'] = df['Order City'].replace('Tokio', 'Tokyo')
df['Order Country'] = df['Order Country'].replace('Singapur', 'Singapore')
df['Order Country'] = df['Order Country'].replace('Repblica Dominicana', 'Dominican Republic')
```
